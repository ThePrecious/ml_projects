## TEMPOGRAM

```python
tempogram_path = Path('/home/ec2-user/SageMaker/course-v3/nbs/personal_projects/Music_Mood_Classification/tempogram')
label_path  = Path('/home/ec2-user/SageMaker/course-v3/nbs/personal_projects/Music_Mood_Classification/input_label.csv')
```

```python
dataT = ImageDataBunch.from_csv(tempogram_path.parent,
                               tempogram_path.name, 
                               csv_labels = str(label_path), suffix='.png',size=224, bs=64)
```

```python
dataT.normalize(imagenet_stats)
```

```python
#dataT.show_batch(rows=3, figsize=(7,6))
```

```python
from  fastai.vision import models

learnT = create_cnn(dataT, models.resnet50, metrics=[error_rate, accuracy])
learnT.lr_find()
learnT.recorder.plot()
```

```python
learnT.fit_one_cycle(25, 5e-2) 
```

```python
learnT.fit_one_cycle(25, max_lr=slice(5e-2, 1e-1)) 
```

```python
learnT.fit_one_cycle(25, max_lr=slice(5e-3, 5e-2)) 
```

```python
learnT.fit_one_cycle(25, 5e-2) 
```

```python
learnT.save('R50_tempogram')
```

```python
# Notes

Default Batch Size
LR (5e-2) Acc: 50%
    
Batch Size = 128 - Cuda out of memory

Batch size = 64
max_lr=slice(5e-2, 1e-1) - Acc 51.5%
max_lr=slice(5e-3, 5e-2) - Acc 50.66%
LR 5e-2  - Acc 51.3
    
    
```
